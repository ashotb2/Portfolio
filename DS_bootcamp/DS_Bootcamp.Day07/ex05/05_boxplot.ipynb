## use the data from the file, read it to a dataframe and make any modification that you may find useful to solve the task 

```python
import pandas as pd

df = pd.read_csv('../data/ab-test.csv')
before = df[df['time'] == 'before']
after = df[df['time'] == 'after']
```

- the figsize is still the same, you can choose whatever fontsize you like
- the color palette should be the same as in the example
- the fontsize of the title is 15
- the width of the box lines is 3, the width of the median lines is 2

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

boxprops = dict(linewidth=3, color='green')
medianprops = dict(linewidth=2, color='red')
meanprops = dict(marker='^', markerfacecolor='green', markersize=6)
whiskerprops = dict(linewidth=1, color='red')
capprops = dict(linewidth=1, color='red')

before.boxplot(column='diff', by='group', ax=axes[1], boxprops=boxprops, medianprops=medianprops, showmeans=True, meanprops=meanprops, whiskerprops=whiskerprops, capprops=capprops)
after.boxplot(column='diff', by='group', ax=axes[0], boxprops=boxprops, medianprops=medianprops, showmeans=True, meanprops=meanprops, whiskerprops=whiskerprops, capprops=capprops)


axes[0].set_title('after')
axes[0].set_xlabel('[group]')
axes[1].set_yticklabels([])
axes[1].set_title('before')
axes[1].set_xlabel('[group]')

y_min, y_max = axes[0].get_ylim()
axes[1].set_ylim(y_min, y_max)

plt.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=16)
plt.show()
```
